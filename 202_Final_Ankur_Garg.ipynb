```python
#Asking for destination they want to go to

def destinationCity(df):
    req_city = input("What city do you want to go to?")
    
    #find if requested city, state, or country is in the database and build df
    req_city_result = df[df.CityDestination.str.contains(req_city)]
    
    #if no result, ask again until valid input
    while len(req_city_result) == 0:
        req_city = input("We dont currently have that destination. What city do you want to go to?")

        #building df if valid
        req_city_result = df[df.CityDestination.str.contains(req_city)]
    
    #creates list of indexes for the 
    req_city_index = req_city_result.index.tolist()

    return req_city, req_city_result, req_city_index

#call function later
# cityReq, cityResult, cityIndex = destinationCity(Basic) 
# cityResult
```

```python
#which rating?

def rating(df):
    
    req_rating = int(input("Above which rating(1-5) would you like to see?"))
    
    while req_rating < 1 or req_rating > 5:
        req_rating = int(input("ERROR! Rating must be between 1 and 5. Above which rating(1-5) would you like to see?"))

    #df of results
    req_rating_result = df.loc[df['Rating'] >= req_rating]

    #List of indexes
    req_rating_index = req_rating_result.index.tolist()

    
    return req_rating, req_rating_result, req_rating_index

#call function
# ratingReq, ratingResult, ratingIndex = rating(Activities)
# print(ratingIndex)
```

```python
# origin city

def originCity(df):
    req_origin = input("What city are you starting at?")
    
    #find if requested origin is in the database and build df
    req_origin_result = df[df.CityOrigin.str.contains(req_origin)]
    
    #if no result, ask again until valid input
    while len(req_origin_result) == 0:
        req_origin = input("We dont currently have any submissions for that origin. What city are you starting at?")

        #building df if valid
        req_origin_result = df[df.CityOrigin.str.contains(req_origin)]
    
    #creates list of indexes for the 
    req_origin_index = req_origin_result.index.tolist()

    return req_origin, req_origin_result, req_origin_index

#call function later
# originReq, originResult, originIndex = originCity(Basic) 
# originIndex
```

```python
#cost range

def cost(df):
    req_cost = int(input("How much are you willing to spend?"))

    #df of results
    req_cost_result = df.loc[df['Cost'] <= req_cost]

    #List of indexes
    req_cost_index = req_cost_result.index.tolist()


    #filter cost by individual categories
    
    
    
    return req_cost, req_cost_result, req_cost_index

#call function
# costReq, costResult, costIndex = cost(Transportation)
# print(costResult)
```

```python
# Transportaiton choices

def transportation(df):
    req_trans = input("Do you prefer (A)driving or (B)flying? Please respond with either (A) or (B).")

    #check for bad input
    while req_trans != "A" and req_trans != "B":
        req_trans = input("Invalid input, please select A or B.")

    # differentiate between car and plane
    if req_trans == "A":
        req_trans = "Car"

    elif req_trans == "B":
        req_trans = "Plane"

    else:
        req_trans = ""
        
    #build df of all flight or car options
    req_trans_result = df[df.TransportationMode.str.contains(req_trans)]

    #creates list of indexes for the 
    req_trans_index = req_trans_result.index.tolist()

    
    #filter by brand
    #car
    if req_trans == "Car":
        print("Would you like to filter by rental company brands or personal vehicle?")
        req_brand = input("Enter No if you dont care, Personal to see POV trips, or the brand you are looking for.")

        #exit if No
        if req_brand == "No":
            return req_trans, req_trans_result, req_trans_index
        else:
            #find if requested city, state, or country is in the database and build df
            req_brand_result = Transportation[Transportation.Car.str.contains(req_brand, na=False)]
            
            #if no result, ask again until valid input
            while len(req_brand_result) == 0:
                print("ERROR or brand is not currently available, please try again")
                req_brand = input("Enter No if you dont care, P to see personal vehicles, or the brand you are looking for.")
                if req_brand == "No":
                    return req_trans, req_trans_result, req_trans_index
                #building df if valid
                req_brand_result = Transportation[Transportation.Car.str.contains(req_brand, na=False)]

            #get index
            req_brand_index = req_brand_result.index.tolist()
            req_brand = req_trans + req_brand

            return req_brand, req_brand_result, req_brand_index
    
    #filter by airline
    if req_trans == "Plane":
        print("Would you like to filter by Airlines?")
        req_brand = input("Enter No if you dont care or the airline you are looking for.")

        #exit if No
        if req_brand == "No":
            return req_trans, req_trans_result, req_trans_index
        else:
            #find if requested city, state, or country is in the database and build df
            req_brand_result = Transportation[Transportation.Airline.str.contains(req_brand, na=False)]
            
            #if no result, ask again until valid input
            while len(req_brand_result) == 0:
                print("ERROR or airline is not currently available, please try again")
                req_brand = input("Enter No if you dont care or the airline you are looking for.")
                if req_brand == "No":
                    return req_trans, req_trans_result, req_trans_index
                #building df if valid
                req_brand_result = Transportation[Transportation.Airline.str.contains(req_brand, na=False)]

            #get index
            req_brand_index = req_brand_result.index.tolist()
            req_brand = req_trans + req_brand
            
            return req_brand, req_brand_result, req_brand_index
            

#call function later
# transReq, transResult, transIndex = transportation() 
# print(transReq)
# print(transIndex)
# print(transResult)
    
```

```python
# What kind of stay?

def stay(df):
    req_stay = input("Do you prefer (A)Hotels or (B)Airbnbs?")

    #check for bad input
    while req_stay != "A" and req_stay != "B":
        req_stay = input("Invalid input, please select A or B.")

    # differentiate between car and plane
    if req_stay == "A":
        req_stay = "Hotel"

    elif req_stay == "B":
        req_stay = "Airbnb"

    else:
        req_stay = ""
        
    #build df of all flight or car options
    req_stay_result = df[df.Stay.str.contains(req_stay)]

    #creates list of indexes for the 
    req_stay_index = req_stay_result.index.tolist()

    
    #filter by brand
    #car
    if req_stay == "Hotel":
        print("Would you like to filter by hotel chains?")
        req_brand = input("Enter No if you dont care or the chain you are looking for.")

        #exit if No
        if req_brand == "No":
            return req_stay, req_stay_result, req_stay_index
        else:
            #find if requested city, state, or country is in the database and build df
            req_brand_result = Stay[Stay.Hotel.str.contains(req_brand, na=False)]
            
            #if no result, ask again until valid input
            while len(req_brand_result) == 0:
                print("ERROR or brand is not currently available, please try again")
                req_brand = input("Enter No if you dont care, P to see personal vehicles, or the brand you are looking for.")
                if req_brand == "No":
                    return req_stay, req_stay_result, req_stay_index
                #building df if valid
                req_brand_result = Stay[Stay.Hotel.str.contains(req_brand, na=False)]
            
            #get index
            req_brand_index = req_brand_result.index.tolist()
            req_brand = req_stay + req_brand

            return req_brand, req_brand_result, req_brand_index
    else:
        return req_stay, req_stay_result, req_stay_index


#call function later
# stayReq, stayResult, stayIndex = stay() 
# print(stayReq)
# print(stayIndex)
# print(stayResult)
```

```python
# Seasons

def season(df):
    req_season = input("Which season are you most interested in: (A)Winter, (B)Spring, (C)Summer, (D)Fall?")

    #check for bad input
    while req_season != "A" and req_season != "B" and req_season != "C" and req_season != "D":
        req_season = input("Invalid input, please select A, B, C, or D.")

    # differentiate between seasons by month
    if req_season == "A":
        req_season = [12, 1, 2]
    elif req_season == "B":
        req_season = [3, 4, 5]
    elif req_season == "C":
        req_season = [6, 7, 8]
    else:
        req_season = [9, 10, 11]


    req_season_result = pd.DataFrame()

    #get results in one df
    for i in req_season:
        req_season_result_infor = df[df['Departure'].dt.month == i]
        frames = [req_season_result, req_season_result_infor]
        req_season_result = pd.concat(frames)

    #get index
    req_season_index = req_season_result.index.tolist()

    return req_season, req_season_result, req_season_index

#call function later
# seasonReq, seasonResult, seasonIndex = season() 
# print(seasonReq)
# print(seasonIndex)
# print(seasonResult)
```

## DRIVER CODE
each function returns:

req_ _result - type: df    
req_ _index - type: list    

available functions:

stay(df)    stayReq, stayResult, stayIndex = stay()

```python
#Simple request for a single filtered list of itineraries based on chosen filters

outputList = []
transReq = ""
stayReq = ""
print("Which filters would you like to use?")
driver_req = input("(A)Destination City, (B)Origin City, (C)Ratings, (D)Cost, (E)Mode of Transport, (F)Type of Stay, (G)Season")


if "A" in driver_req: # city destination
    cityReq, cityResult, cityIndex = destinationCity(Basic)
    outputList.append(cityIndex)
    
if "B" in driver_req: # origin city
    originReq, originResult, originIndex = originCity(Basic)
    outputList.append(originIndex)
    
if "C" in driver_req: # Ratings
    print("Which of these do you want to filter ratings for?")
    ratingQ = input("(A)Overall, (B)Transport, (C)Stay, (D)Activities")
    if "A" in ratingQ:
        print("Overall Rating")
        ratingReqOverall, ratingResultOverall, ratingIndexOverall = rating(Basic)
        outputList.append(ratingIndexOverall)
    if "B" in ratingQ:
        print("Transport Rating")
        ratingReqTransport, ratingResultTransport, ratingIndexTransport = rating(Transportation)
        outputList.append(ratingIndexTransport)
    if "C" in ratingQ:
        print("Stay Rating")
        ratingReqStay, ratingResultStay, ratingIndexStay = rating(Stay)
        outputList.append(ratingIndexStay)
    if "D" in ratingQ:
        print("Activities cost")
        costReqActivities, costResultActivities, costIndexActivities = cost(Activities)
        outputList.append(costIndexActivities)

if "D" in driver_req: # Cost
    print("Which of these do you want to filter Costs for?")
    costQ = input("(A)Overall, (B)Transport, (C)Stay, (D)Activities")
    if "A" in costQ:
        print("Overall Cost")
        costReqOverall, costResultOverall, costIndexOverall = cost(Basic)
        outputList.append(costIndexOverall)
    if "B" in costQ:
        print("Transport Cost")
        costReqTransport, costResultTransport, costIndexTransport = cost(Transportation)
        outputList.append(costIndexTransport)
    if "C" in costQ:
        print("Stay Cost")
        costReqStay, costResultStay, costIndexStay = cost(Stay)
        outputList.append(costIndexStay)
    if "D" in costQ:
        print("Activities Cost")
        costReqActivities, costResultActivities, costIndexActivities = cost(Activities)
        outputList.append(costIndexActivities)

if "E" in driver_req: # Mode of Transport
    transReq, transResult, transIndex = transportation(Basic)
    outputList.append(transIndex)

if "F" in driver_req: # Type of Stay
    stayReq, stayResult, stayIndex = stay(Basic)
    outputList.append(stayIndex)

if "G" in driver_req: # Seasons
    seasonReq, seasonResult, seasonIndex = season(Basic)
    outputList.append(seasonIndex)

#finds the common indexes from all the filters to create the final list
elements_in_all = list(set(outputList[0]).intersection(*outputList[1:]))

#extract info from each page of the excel corresponding with indexes above
basicFin = Basic.iloc[elements_in_all]
transFin = Transportation.iloc[elements_in_all]
stayFin = Stay.iloc[elements_in_all]
activitiesFin = Activities.iloc[elements_in_all]

#hide columns not in use based on filter
if transReq == "Plane":
    transFin = transFin.drop('Car', axis=1)
    transFin = transFin.drop('Distance', axis=1)
if transReq == "Car":
    transFin = transFin.drop('Airline', axis=1)
    transFin = transFin.drop('PlaneClass', axis=1)

if stayReq == "Hotel":
    stayFin = stayFin.drop('Airbnb', axis=1)
if stayReq == "Airbnb":
    stayFin = stayFin.drop('Hotel', axis=1)

#Final Output
print("Here are some sample itineraries from real people!!!")
display(basicFin)
display(transFin)
display(stayFin)
display(activitiesFin)
print("disclaimer: this is not real data, it was randomly generated to display concept")
```
